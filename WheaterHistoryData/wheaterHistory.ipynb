```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)


df['Year'] = df['Formatted Date'].dt.year
df['Month'] = df['Formatted Date'].dt.month
df['Day'] = df['Formatted Date'].dt.day
```

```python
df = df.drop("Formatted Date",axis = 1)
```

```python
objectColumns = ["Summary","Daily Summary","Precip Type"]
for i in objectColumns:
  print(df[i].unique())
#The Daily Summry is unnecessary
df = df.drop("Daily Summary",axis = 1)
objectColumns = ["Summary","Precip Type"]
for i in objectColumns:
  print(df[i].unique())
```

```python
objectColumns = ["Summary","Precip Type"]
for i in objectColumns:
  print(df[i].unique())
```

```python
def simplify_summary(summary):
    summary = summary.lower()
    if 'rain' in summary or 'drizzle' in summary:
        return 'Rain'
    elif 'cloudy' in summary or 'overcast' in summary:
        return 'Cloudy'
    elif 'fog' in summary:
        return 'Foggy'
    elif 'clear' in summary:
        return 'Clear'
    elif 'wind' in summary or 'breezy' in summary:
        return 'Windy'
    else:
        return 'Other'
df["Summary"] = df["Summary"].apply(simplify_summary)
```

```python
for i in objectColumns:
  print(df[i].unique())
```

```python
df = pd.get_dummies(df, columns=['Summary',"Precip Type"],dtype = "int64")
```

```python
df = df.drop(["Year","Day","Pressure (millibars)","Summary_Other","Summary_Rain","Summary_Windy","Summary_Other"],axis = 1)
```

Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop("Temperature (C)",axis = 1)
Y = df["Temperature (C)"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.25,random_state=42)
```

# **DecisionTreeRegressor**

```python

model = DecisionTreeRegressor()
model.fit(x_train,y_train)
y_preds = model.predict(x_test)
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)

print("R² train:", r2_score(y_train, y_pred_train))
print("R² test:", r2_score(y_test, y_pred_test))
print("MAE Score = ",mean_absolute_error(y_test,y_preds))
print("MSE Score = ",mean_squared_error(y_test,y_preds))
sns.scatterplot(x = y_test,y = y_preds)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# **SVR**

```python
model = SVR()
model.fit(x_train,y_train)
y_preds = model.predict(x_test)
print("R2 Score = ",r2_score(y_test,y_preds))
print("MAE Score = ",mean_absolute_error(y_test,y_preds))
print("MSE Score = ",mean_squared_error(y_test,y_preds))
sns.scatterplot(x = y_test,y = y_preds)
```

# **KNN**

```python
model = KNeighborsRegressor()
model.fit(x_train,y_train)
y_preds = model.predict(x_test)
print("R2 Score = ",r2_score(y_test,y_preds))
print("MAE Score = ",mean_absolute_error(y_test,y_preds))
print("MSE Score = ",mean_squared_error(y_test,y_preds))
sns.scatterplot(x = y_test,y = y_preds)
```

# **Linear Regression**

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_preds = model.predict(x_test)
print("R2 Score = ",r2_score(y_test,y_preds))
print("MAE Score = ",mean_absolute_error(y_test,y_preds))
print("MSE Score = ",mean_squared_error(y_test,y_preds))
sns.scatterplot(x = y_test,y = y_preds)
```
